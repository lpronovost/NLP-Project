```python
import gensim
import pprint
from gensim import corpora
from gensim.utils import simple_preprocess
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
doc_list = []
count = 0
for tf in txtfiles:
    path = 'training_20180910/'+tf
    f = open(path)
    text = f.read()
    text = re.sub("[^A-Za-z]+", " ", text)
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if not w.lower() in stopwords.words("english")]
    text = " ".join(tokens)
    text = text.lower().strip()
    doc_list.append(text)
    count+=1
    print(count, "files cleaned")

doc_list
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# import other required libs
import pandas as pd
import numpy as np

vectorizer = TfidfVectorizer(sublinear_tf=True, min_df=5, max_df=0.95)
# fit_transform applies TF-IDF to clean texts - we save the array of vectors in X
X = vectorizer.fit_transform(doc_list)
pca = PCA(n_components=2, random_state=42)
pca_x = pca.fit_transform(X.toarray())
```

```python
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

```python
sns.set(rc={'figure.figsize':(10,8)})
import pandas as pd
import seaborn as sns
clusters = np.array(KMeans(n_clusters=8).fit_predict(pca_x))
graph_dat = pd.DataFrame(pca_x, columns=['x','y'])
graph_dat['cluster'] = clusters
sns.scatterplot(data=graph_dat, x = 'x', y = 'y', hue = 'cluster', palette='viridis')
```
