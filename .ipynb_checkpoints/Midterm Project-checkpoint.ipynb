```python
import os
import re
from collections import Counter, defaultdict
from nltk import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import scispacy
from spacy import displacy 
import en_ner_bc5cdr_md 
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import time
```

```python
files = os.listdir('training_20180910')
txtfiles = []
for f in files:
    if f.endswith('.txt'):
        txtfiles.append(f)
        
nltk.download('stopwords')
doc_list = []
for tf in txtfiles:
    path = 'training_20180910/'+tf
    f = open(path)
    text = f.read()
    text = re.sub("[^A-Za-z]+", " ", text)
    tokens = nltk.word_tokenize(text)
    tokens = [w for w in tokens if not w.lower() in stopwords.words("english")]
    text = " ".join(tokens)
    text = text.lower().strip()
    doc_list.append(text)

vectorizer = TfidfVectorizer(sublinear_tf=True, min_df=5, max_df=0.95)
# fit_transform applies TF-IDF to clean texts - we save the array of vectors in X
X = vectorizer.fit_transform(doc_list)
pca = PCA(n_components=2, random_state=42)
pca_x = pca.fit_transform(X.toarray())
```

```python
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.axvline(6, color='r')
plt.axvline(8, color='r')
plt.title('Elbow curve')
```

```python
sns.set(rc={'figure.figsize':(10,8)})
clusters = np.array(KMeans(n_clusters=8).fit_predict(pca_x))
graph_dat = pd.DataFrame(pca_x, columns=['x','y'])
graph_dat['cluster'] = clusters
sns.scatterplot(data=graph_dat, x = 'x', y = 'y', hue = 'cluster', palette='viridis')
```

```python
# find_sections takes in a file and a dictionary of relating to section counts. 
# Using the typical format of a doctors note, this section extracts and returns the chief complaint, HPI, and discharge diagnosis of each note
# The NER work better when the file was in its individual format, hence not removing any special characters.

def find_section(f, counts):
    dd_section = ""
    cc_section = ""
    hpi_section = ""
    dd_done = False
    dd_looking = False
    cc_done = False
    cc_looking = False
    hpi_done = False
    hpi_looking = False

    
    for line in f:
        line = line.lower()
        
        if re.search(r'discharge diagnosis',line) != None: # heading of this section
            dd_looking = True
            counts['dd_l']+=1
        elif dd_looking and not dd_done:
            if re.search(r'discharge',line) != None: # the section follow is 
                dd_done = True
                counts['dd_d']+=1
            else:
                dd_section = dd_section+line
        
        if re.search(r'chief complaint',line) != None:
            cc_looking = True
            counts['cc_l']+=1
        elif cc_looking and not cc_done:
            if re.search(r'major surgical or invasive procedure',line) != None or re.search(r'history of present illness',line) != None:
                cc_done = True
                counts['cc_d']+=1
            else:
                cc_section = cc_section+line
                
        elif re.search(r'history of present illness',line) != None or re.search(r'history of the present illness',line) != None:
            hpi_looking = True
            counts['hpi_l']+=1
        elif hpi_looking and not hpi_done:
            if re.search(r'past medical history',line) != None or re.search(r'review of systems',line) != None:
                hpi_done = True
                counts['hpi_d']+=1
            else:
                hpi_section = hpi_section+line
                
    return dd_section, cc_section, hpi_section, counts
```

```python
# Clean_diag takes in a diagnosis extracted from the NER
# This diagnosis has sepcial characters removed, is lemmatized, and then is cleaned for common errors I found after running the model a number of times
# changes is a dictionary of common occurances that need to be correct
# contains is a list of common occurances that can be simplified to their base root
# remove is a list common extra medical language that adds too much complexity

def clean_diag(diag):
    
    changes = {
    'htn':'hypertension',
    'atrial fibrilation':'atrial fibrillation',
    'copd': 'obstructive pulmonary disease', 
    'hypotensive':'hypotension',
    'gerd':'gastroesophageal reflux disease',
    'abdominal':'abdominal pain',
    'aaa':'abdominal aortic aneurysm',
    'dvt':'deep vein thrombosis',
    'aml':'leukemia',
    'emesis':'vomiting',
    'migraine':'headache',
    'chill':'chills',
    'hypertensive':'hypertension',
    'hypothyroid':'hypothyroidism'} #common occurances that needed to be changed
    
    contains = ['diabetes','arthritis', 'anemia', 'gerd','pneumonia',
                'abdominal','aaa','vomiting', 'leukemia', 's p', 'adenocarcinoma'] #common words that pertaining to overly complicated or wordy diagnoses
    remove = ['chronic','end stage', 'acute', 'traumatic', 'herpin induced', 'mid '] #prefixes that can be removed
    lemmatizer = WordNetLemmatizer()
    diag = re.sub("[^A-Za-z]+", " ", diag).strip()
    diag = lemmatizer.lemmatize(diag)
    
    for word in remove:
        diag = diag.replace(word,"").strip()
    for word in contains:
        if re.search(word,diag) != None:
            diag = word
    if diag in changes.keys():
        diag = changes[diag]
    return diag
```

```python
# commmon "diseases" that should not be considered in the corpus
ignore = ['allergies', 'pain', 's p', 'treatment work up', 'history', 'allergy', 'artery to common femoral artery bypass and femoral femoral cross over graft were performed']
# common symptoms that shouldn't be considered medical conditions -- ignoring psychiactric conditions as well (this was mentioned in the example of a medical condition)
common_symp = ['nausea', 'vomiting', 'fever', 'bleeding', 'abdominal pain', 'diarrhea', 'headache', 'depression', 'anxiety', 'bipolar disorder']
```

```python
# make_condition_dicts takes in a NER document (doc), a file name (tf), a diagnosis counter (diag), and a dictionary of diagnoses to files (files)
# this method updates and returns the passed counter and dictionary with the diseases from the NER and given file name
# each disease from the NER doc is passed to clean_diag
# used on Chief Complaint and Discharge Diagnosis

def make_condition_dicts(doc, tf, diags, files):
    for word in doc.ents:
        if word.label_ == 'DISEASE':
            disease = clean_diag(word.text)
            if disease not in ignore and disease not in common_symp:
                diags[disease]+=1
                files[disease].add(tf)
    return diags, files
```

```python
# make_words takes in a NER document (doc) and returns a list of cleaned diagnoses from the NER
# used on HPI and the entire note

def make_words(doc):
    words = []
    for word in doc.ents:
        if word.label_ == 'DISEASE':
            disease = clean_diag(word.text)
            if disease not in ignore:
                words.append(disease)
    return words
```

```python
t0 = time.time()

nlp = en_ner_bc5cdr_md.load()
hpi_data = pd.DataFrame(columns=['file','note']) # data frame containg the file name and conditions found in its HPI
note_data = pd.DataFrame(columns=['file','note']) # data frame containg the file name and conditions found in the entire note

dd_conditions = Counter() #counts the number occurrences of each disease from the discharge diagnosis section
diagnosis_file = defaultdict(lambda: set()) #keys: diagnoses; values: list of files with diagnosis key in discharge diagnosis section 
cc_conditions = Counter() #counts the number occurrences of each disease from the chief complaint section
complaint_file = defaultdict(lambda: set()) #keys: diagnoses; values: list of files with the diagnosis key in chief complaint section 

section_counts = Counter()

for tf in txtfiles:
    added = False
    hpi_words = []
    note_words = []
    path = 'training_20180910/'+tf
    f = open(path, 'r')
    
    dd_section, cc_section, hpi_section, section_counts = find_section(f, section_counts)
    discharge_doc = nlp(dd_section) # generating an NER doc of the discharge diagnosis from note tf
    complaint_doc = nlp(cc_section) # generating an NER doc of the chief complaint from note tf
    history_doc = nlp(hpi_section) # generating an NER doc of the HPI from note tf
    f = open(path, 'r') # re-opening the file
    note_doc = nlp(f.read().lower()) # generating an NER doc of the entire note tf
    
    dd_conditions, diagnosis_file = make_condition_dicts(discharge_doc, tf, dd_conditions, diagnosis_file) #updating discharge diagnosis condition counter and diagnosis to file dictionary
    cc_conditions, complaint_file = make_condition_dicts(complaint_doc, tf, cc_conditions, complaint_file) #updating chief complaint condition counter and diagnosis to file dictionary
    hpi_words = make_words(history_doc) # extracting diseases from HPI of note tf
    note_words = make_words(note_doc) # extracting diseases from entire note tf
    
    if len(hpi_words) > 0:        
        row = {'file': tf, 'note': hpi_words}
        hpi_data = hpi_data.append(row, ignore_index = True) # updating HPI data frame 
    if len(note_words) > 0:        
        row = {'file': tf, 'note': note_words}
        note_data = note_data.append(row, ignore_index = True) # updating note data frame

t1 = time.time()
print(f"Run time: {round((t1-t0)/60,4)} minutes")
```

```python
def plot_counter(conditions, n):
    xs = []
    ys = []
    for group in conditions.most_common(n):
        xs.append(group[0])
        ys.append(group[1])
    plt.figure(figsize=(20,10))
    plt.xticks(rotation = 90)
    plt.xlabel( "Condition" , size = 12 )
    plt.ylabel( "Condition Count" , size = 12 )
    sns.barplot(x = xs, y = ys)
```

```python
# get_corpora takes in a diagnosis and data frame representing the BagOfDiagnoses of each note 
# using the diagnosis to file dictionary of BagOfDiagnoses of each file, generates a BagOfDiagnoses for the given Diagnosis
# creates a corpora of word ids and word frequency
# returns the corpus, dictionary of id to words, and the complete BagOfDiagnoses for the given diagnosis

def get_copora(diagnosis, data):
    file_names = list(diagnosis_file[diagnosis])
    notes = []
    for file in file_names:
        if file in list(data['file']):
            note = data[data['file'] == file]['note'].values[0]
            notes.append(note)
    id2word = corpora.Dictionary(notes)
    corpus = [id2word.doc2bow(text) for text in notes]
    return corpus, id2word, notes
```

```python
# get_probs takes the resulting topics of a LDA model (model) and the diagnosis pertaining to the copora the model was fit to
# creates and returns a dictionary of topic frequencys

def get_probs(model, diagnosis):
    diagnosis_freq = defaultdict(lambda: 0)
    for i in range(0,len(model)):
        res = model[i][1].split("+")
        for group in res:
            prob = float(group.split('*')[0].strip())
            word = group.split('*')[1].strip()[1:-1]
            if word != diagnosis:
                diagnosis_freq[word]+=prob
    return diagnosis_freq
```

```python
def generate_conditions(condition_dict, n, data, default=True, a=0, d=0):
    t0 = time.time()
    diagnoses = condition_dict.most_common(n) #gets the top n most common diagnoses
    final_conditions = defaultdict(lambda: [])
    coherence_lda = 0
    count = 0
    for diagnosis in diagnoses:
        corpus, id2word, text_data = get_copora(diagnosis[0], data) # generating a corpus for a diagnosis
        if default:
            lda_model = gensim.models.LdaMulticore(corpus=corpus,id2word=id2word,num_topics=10, passes=100, random_state=42) #used for initail model
        else:
            lda_model = gensim.models.LdaMulticore(corpus=corpus,id2word=id2word,num_topics=10, passes=100, alpha=a, decay=d, random_state=42) #used in hpyertuning
        model = lda_model.print_topics()
        diagnosis_freq = get_probs(model, diagnosis[0]) #generating frequency of topics in diagnosis corpora
        vals = list(diagnosis_freq.values())
        average_freq = np.mean(vals)
        freq_std = np.std(vals)
        for item in diagnosis_freq.items():
            if item[1] >= average_freq+1.5*freq_std: 
                final_conditions[diagnosis[0]].append(item[0]) #adding topic as underlying condition of the given diagnosis
        coherence_model_lda = CoherenceModel(model=lda_model, texts=text_data, dictionary=id2word, coherence='c_v') 
        cohere = coherence_model_lda.get_coherence() #had to name the coherence value "cohere" 
        if not np.isnan(cohere):
            coherence_lda += cohere
            count+=1
    avgc = coherence_lda/count #average coherence across all diagnoses
    print(f"Average Coherence: {avgc}")
    t1 = time.time()
    print(f"Run time: {round((t1-t0)/60,4)} minutes")
    return final_conditions, avgc
```

```python
def count_vals(res):
    res_count = Counter()
    for val in res.values():
        for condition in val:
            res_count[condition]+=1
    return res_count
```

```python
def compute_coherence_values(condition_dict, n, data, decays):
    one_coherence_values = []
    two_coherence_values = []
    three_coherence_values = []
    for decay in decays:
        print(f"Alpha 0.1 model. Decay value: {decay}")
        model_one, one_coherence = generate_conditions(condition_dict, n, data, False,0.1,decay)
        print(f"\nAlpha 0.5 model. Decay value: {decay}")
        model_two, two_coherence = generate_conditions(condition_dict, n, data, False,0.5,decay)
        print(f"\nAlpha 1.0 model. Decay value: {decay}")
        model_three, three_coherence = generate_conditions(condition_dict, n, data, False,1.0,decay)
        print("\n")               
        one_coherence_values.append(one_coherence)
        two_coherence_values.append(two_coherence)
        three_coherence_values.append(three_coherence)
        
    return one_coherence_values, two_coherence_values, three_coherence_values

decays = [0.6, 0.8, 1.0]
one, two, three = compute_coherence_values(dd_conditions, 20, hpi_data, decays)
print("Base model")
base_model, base_c = generate_conditions(dd_conditions, 20, hpi_data)
```

```python
plt.plot(decays, one, label='alhpa = 0.1')
plt.plot(decays, two, label='alpha = 0.5')
plt.plot(decays, three, label='alpha = 1')
plt.axhline(base_c, color='r')
plt.xlabel("Decay")
plt.ylabel("Average Coherence score")
plt.legend(loc='best')
plt.show()
```
